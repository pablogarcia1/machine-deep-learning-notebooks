```python
from sklearn import metrics
def get_accuracy(X_train, X_test, y_train, y_test, model):
    return  {"test Accuracy":metrics.accuracy_score(y_test, model.predict(X_test)),"trian Accuracy": metrics.accuracy_score(y_train, model.predict(X_train))}
```

With this we can observe that GradeClass it´s unbalanced so we will use an StratifiedSplit to separate our variables

Then we can Split our Variables

```python
from sklearn.model_selection import train_test_split, ShuffleSplit, GridSearchCV
X_columns_drop = ["GradeClass","StudentID", "GPA"]
X = df.drop(columns=X_columns_drop) #Predictive values    
y = df["GradeClass"] #Objective values

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, stratify=y, random_state= 1)
```

```python
X
```

We can observe the difference now using an stratified spit

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

We can apply now our first model Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV


model = RandomForestClassifier()
```

```python
param_grid = {'n_estimators': [n for n in range(20)],
             'max_depth' : [n for n in range(10) ],
             'max_features':["auto", "sqrt", "log2"],
              'criterion': ["gini","entropy","log_loss"]
}


RF_model = GridSearchCV(estimator=model, param_grid= param_grid, scoring="accuracy")

RF_model.fit(X_train,y_train)
```

```python
RF_model.best_score_
```

```python
RF_model.best_params_
```

```python
RF_model.best_estimator_
```

```python
get_accuracy(X_train, X_test, y_train, y_test, RF_model.best_estimator_)
```

```python
best_RF_model = RF_model.best_estimator_

y_pred = best_RF_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm_RF = confusion_matrix(y_test, y_pred, labels= best_RF_model.classes_)
disp = ConfusionMatrixDisplay(cm_RF, display_labels= best_RF_model.classes_)
disp.plot(cmap= "Blues")
plt.show()
```

We can try now different models to see the differences

```python
from sklearn.ensemble import GradientBoostingClassifier

param_grid_GB = {
    'n_estimators': [50, 100, 150],  # Eliminé 200
    'learning_rate': [0.05, 0.1],  # Eliminé 0.01 y 0.5
    'max_depth': [3, 4, 6]  # Eliminé 8 y 10
}

model = GradientBoostingClassifier()
GB_model = GridSearchCV(model, param_grid_GB, scoring="accuracy")

GB_model.fit(X_train,y_train)
```

```python
GB_model.best_score_
```

```python
get_accuracy(X_train, X_test, y_train, y_test, GB_model.best_estimator_)
```

```python
best_GB_model = GB_model.best_estimator_
y_pred = best_GB_model.predict(X_test)
```

```python
cm_GB = confusion_matrix(y_test, y_pred, labels= best_GB_model.classes_)
disp = ConfusionMatrixDisplay(cm_GB, display_labels= best_RF_model.classes_)
disp.plot(cmap= "Blues")
plt.show()
```

Data Scaling 


```python
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)  # Escalar las características
```

We split our data 

```python
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=1, stratify=y)
```

```python
from sklearn.svm import SVC

model = SVC()
param_grid_SVC = {
    'C': [0.1, 1, 10],  # Menos valores para C
    'kernel': ['linear', 'rbf'],  # Reducimos el número de kernels
    'gamma': ['scale', 'auto'],  # Menos valores de gamma
    'degree': [2, 3],  # Solo para 'poly' kernel, pero solo probamos 2 y 3
    'shrinking': [True, False],  # Lo dejamos igual
    'tol': [1e-4, 1e-3],  # Dos valores de tolerancia
    'max_iter': [-1, 1000]  # Menos valores para el número de iteraciones
}
SVC_model = GridSearchCV(model, param_grid_SVC, scoring="accuracy", n_jobs=-1)

SVC_model.fit(X_train_scaled,y_train)
```

```python
SVC_model.best_score_
```

```python
SVC_model.best_params_
```

Trying this model we can notice the lineal relation between the varables


```python
best_SVC_model = SVC_model.best_estimator_
y_pred = best_SVC_model.predict(X_test_scaled)


cm_SVC = confusion_matrix(y_test, y_pred, labels= best_SVC_model.classes_)
disp = ConfusionMatrixDisplay(cm_SVC, display_labels= best_SVC_model.classes_)
disp.plot(cmap= "Blues")
plt.show()

```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()

param_grid_knn = {
    'n_neighbors': [n for n in range(50)],  # Número de vecinos a considerar
    'weights': ['uniform', 'distance'],  # Peso de los vecinos: uniforme o ponderado por distancia
    'metric': ['euclidean', 'manhattan', 'minkowski'],  # Métrica de distancia (se puede usar 'minkowski' como opción general)
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],  # Algoritmo de búsqueda de vecinos más cercanos
    'leaf_size': [10, 20, 30, 40],  # Tamaño de la hoja para los árboles Ball Tree y KD Tree
    'p': [1, 2],  # Poder de la distancia (1 para Manhattan, 2 para Euclidean)
}


KNN_model = GridSearchCV(model,param_grid_knn,n_jobs=-1, scoring="accuracy")
KNN_model.fit(X_train_scaled,y_train)
```

Now that we got our best model we can find analytic results using feature importance

```python
from sklearn.inspection import permutation_importance
import numpy as np

perm_importance = permutation_importance(best_SVC_model, X_test, y_test, n_repeats=10, random_state=42, n_jobs=-1)
```

```python
# Obtener importancia y nombres de features
sorted_idx = np.argsort(perm_importance.importances_mean)[::-1]  # Ordenar de mayor a menor

# Graficar los resultados
plt.figure(figsize=(10, 5))
plt.barh(np.array(X_test.columns)[sorted_idx], perm_importance.importances_mean[sorted_idx])
plt.xlabel("Permutation Importance")
plt.title("Feature Importance - SVC")
plt.gca().invert_yaxis()  # Para que la más importante esté arriba
plt.show()
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(np.array(X_test.columns)[sorted_idx], perm_importance.importances_mean[sorted_idx], color='dodgerblue')

# Añadir etiquetas y título
plt.xlabel("Features")
plt.ylabel("Permutation Importance")
plt.title("Feature Importance - SVC")
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas de las features para mayor legibilidad
plt.tight_layout()  # Ajustar el diseño para que

# Mostrar el gráfico
plt.show()
```
